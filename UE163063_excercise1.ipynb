```
#Q11
def swap(a,b):
  print(f"before a, b = {a} {b}")
  a = a + b
  b = a - b
  a = a - b
  print(f"After a, b = {a} {b}")
swap(5,6)
```

```
#Q14
def monthly_pay(hours, pay_rate, number_weeks):
    pay = hours*pay_rate*number_weeks
    print(f"Number of hours worked in a week:{hours}")
    print(f"Pay rate per hour:{pay_rate}")
    print(f"Number of weeks:{number_weeks}")
    print(f"total pay:{pay}")
monthly_pay(40, 400, 4)
```

```
#Q15c
def prime(n):
    flag = False
    for i in range(2,n):
        if (n%i==0):
            flag = True
            break
    if (flag):
        print("Not prime")
    else : print("Prime")
prime(int(input("Enter a number:")))
```

```
#Q15d
def fibonacci(n):
    a ,b = 0, 1
    print(a)
    print(b)
    for i in range(2,n):
        c = a+b
        a = b
        b = c
        print(c)

fibonacci(int(input("Enter n:")))
```

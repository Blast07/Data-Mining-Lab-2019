```
def punctuation(word):
    punc = ['.', ',', '!', '?']
    for ch in punc:
      if ch in word:
         return ch
    return False

def word_scramble(string_of_words):
  words = list()
  words = string_of_words.split()
  count = 0
  for word in words:
    if len(word)>3:
      ch = punctuation(word)
      if ch:
        if ch == word[-1]:
          words[count] = scramble(word[:-1])+word[-1]
      else:
        words[count] = scramble(word)
    count+=1
  return ' '.join(words) 

    
```

```
test1="this is a word scrambler program!"
test2="Where is the  abc.doc, xyz.html file?"
print(word_scramble(test1))
print(word_scramble(test2))
```
